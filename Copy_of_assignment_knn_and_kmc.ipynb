**Q5.** This question is a case study for $k$ means clustering.

1. Load the `airbnb_hw.csv` data. Clean `Price` along with `Beds`, `Number of Reviews`, and `Review Scores Rating`.
2. Maxmin normalize the data and remove any `nan`'s (`KMeans` from `sklearn` doesn't accept `nan` input).
3. Use `sklearn`'s `KMeans` module to cluster the data by `Beds`, `Number of Reviews`, and `Review Scores Rating` for `k=6`.
4. Use `seaborn`'s `.pairplot()` to make a grid of scatterplots that show how the clustering is carried out in multiple dimensions.
5. Use `.groupby` and `.describe` to compute the average price for each cluster. Which clusters have the highest rental prices?
6. Use a scree plot to pick the number of clusters and repeat steps 4 and 5.

````python
df3 = pd.read_csv('./assignment3/data/airbnb_hw.csv')
df3 = df3.dropna(subset=['Price', 'Beds', 'Number Of Reviews','Review Scores Rating'])
df3['Price'] = df3['Price'].str.replace('[\$,]', '', regex=True).astype(float)
df3['Beds'] = df3['Beds'].astype(int) # making sure there are only ints for number of beds
max_beds = 10
min_beds = 1
df3['Beds'] = df3['Beds'].apply(lambda x: round(x))
df3['Beds'] = df3['Beds'].clip(lower=min_beds, upper=max_beds).astype(int) #making sure there are no inputs that don't make sense
def onlyDigit(x): #makes sure beds only have ints
   if str(x).isdigit():
     return int(x)
   else:
          return None

df3['Beds'] = df3['Beds'].apply(onlyDigit)
df3['Number Of Reviews'] = df3['Number Of Reviews'].astype(int)
df3['Number Of Reviews'] = df3['Number Of Reviews'].apply(lambda x: round(x))
df3['Review Scores Rating'] = df3['Review Scores Rating'].astype(float)


````

````python
max_Price = df3['Price'].max()
max_Beds = df3['Beds'].max()
max_Number_Of_Reviews = df3['Number Of Reviews'].max()
max_Review_Scores_Rating = df3['Review Scores Rating'].max()
df3['normalizedPrice'] = (df3['Price'] - 0) / max_Price
df3['normalizedBeds'] = (df3['Beds'] - 0) / max_Beds
df3['normalizedNumOfReviews'] = (df3['Number Of Reviews'] - 0) / max_Number_Of_Reviews
df3['normalizedReviewScoresRating'] = (df3['Review Scores Rating'] - 0) / max_Review_Scores_Rating
````

````python
from sklearn.cluster import KMeans
cluster = df3[['Beds', 'Number Of Reviews', 'Review Scores Rating']]
kmeans = KMeans(n_clusters=6, random_state=100)
kmeans.fit(cluster)
cluster_labels = kmeans.labels_
df3['Cluster'] = cluster_labels
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style ="ticks")
sns.pairplot(df3, hue ='Cluster', palette ='Set1')
plt.show()
````

````python
cluster_price = df3.groupby('Cluster')['Price'].mean()
print(cluster_price) #Cluster 2 and 3 has the highest retail prices
````

````python
clusters = df3[['Beds', 'Number Of Reviews', 'Review Scores Rating']]
eginvalues = []
k_values = range(1, 7)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(clusters)
    eginvalues.append(kmeans.inertia_)
````

````python
plt.figure(figsize=(8, 6))
plt.plot(k_values, eginvalues, marker='o')
plt.xlabel('Components')
plt.ylabel('Eginvalue')
plt.title('Scree Plot')
plt.grid()
plt.show()
````
